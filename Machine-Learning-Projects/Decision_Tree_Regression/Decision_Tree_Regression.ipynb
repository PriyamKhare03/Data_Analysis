```python

numberic_col = df.select_dtypes(include ="number").columns.tolist()
categories_bol = df.select_dtypes(include=["bool"]).columns.tolist()
# df[categories[0]].value_counts()

df[categories_bol]  = df[categories_bol].astype(int)
#categories = df.select_dtypes(include=["object","bool"]).columns.tolist()
#print(categories)
categories_col = df.select_dtypes(exclude="number").columns.tolist()

df = pd.get_dummies(df,columns=categories_col,drop_first=True)
df
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
df[numberic_col] = scaler.fit_transform(df[numberic_col])
df[numberic_col]
df
```

```python
X_col = "price"
for col in df.select_dtypes(include=["int64","float64"]).columns:
  if col != X_col:
    df.plot(kind="scatter",x=X_col,y=col,figsize=(10,10),title=f"{X_col} vs {col}")
    plt.show()
```

```python
x = df.drop("price",axis=1)
y = df["price"]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.tree import DecisionTreeRegressor
model = DecisionTreeRegressor(random_state=42,max_depth=8,min_samples_leaf=2,min_samples_split=4,min_impurity_decrease=0.001)
```

```python
model.fit(x_train,y_train)
```

```python
from sklearn.metrics import r2_score,mean_squared_error
y_pred = model.predict(x_test)
mse = mean_squared_error(y_pred,y_test)
r2 = r2_score(y_pred,y_test)
print(f"Error is {mse}")
print(f"r2 score is {r2}")
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(7,5))
plt.scatter(y_test, y_pred, alpha=0.6)
plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)],"r--", lw=1)
plt.xlabel("Actual Price")
plt.ylabel("Predicted Price")
plt.title("Actual vs Predicted Car Prices")
plt.show()
```
